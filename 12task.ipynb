```python
def analyze_string(s):
    if s == "":
        print("Empty string")
        return

    print("Length:", len(s))
    print("Reverse:", s[::-1])

    vowels = "aeiouAEIOU"
    count = 0
    for ch in s:
        if ch in vowels:
            count += 1
    print("Vowels:", count)

    for i in range(len(s)):
        print(s[i], "Positive:", i, "Negative:", i-len(s))

text = input("Enter a string: ")
analyze_string(text)
```

```python
def manage_marks():
    marks = []

    while len(marks) < 5:
        try:
            m = float(input("Enter mark: "))
            marks.append(m)
        except ValueError:
            print("Enter valid number")

    print("Average:", sum(marks)/len(marks))
    print("Highest:", max(marks))
    print("Lowest:", min(marks))

    marks.sort(reverse=True)
    print("Sorted:", marks)

manage_marks()
```

```python
class Student:
    def __init__(self, name, roll):
        self.name = name
        self.roll = roll
        self.marks = []

    def add_mark(self, mark):
        if 0 <= mark <= 100:
            self.marks.append(mark)
        else:
            print("Invalid mark")

    def get_average(self):
        return sum(self.marks) / len(self.marks)

    def display(self):
        print(self.name, self.roll, self.marks)
        print("Average:", self.get_average())

s = Student(input("Name: "), input("Roll: "))

for i in range(3):
    s.add_mark(int(input("Enter mark: ")))

s.display()
```

```python
students = {}

while True:
    print("\n1.Add\n2.Search\n3.Display\n4.Exit")
    ch = input("Choice: ")

    if ch == "1":
        roll = input("Roll: ")
        name = input("Name: ")
        age = input("Age: ")
        city = input("City: ")
        students[roll] = {"Name": name, "Age": age, "City": city}

    elif ch == "2":
        roll = input("Roll: ")
        print(students.get(roll, "Not Found"))

    elif ch == "3":
        print(students)

    elif ch == "4":
        break
```

```python
import random
import math

try:
    s = set()

    for i in range(10):
        n = int(input("Enter number: "))
        s.add(n)

    t = tuple(s)

    print("Tuple:", t)
    print("3 Random Numbers:", random.sample(t, min(3, len(t))))
    print("Square Root:", math.sqrt(sum(t)))

except Exception as e:
    print("Error:", e)
```

```python
square = lambda x: x * x

numbers = list(range(1, 21))
squares = list(map(square, numbers))

print("Even Squares:")
for i in squares:
    if i % 2 == 0:
        print(i)
```

```python
class Employee:
    def __init__(self, emp_id, name, dept, salary):
        self.emp_id = emp_id
        self.name = name
        self.details = (dept, salary)

    def show_details(self):
        print(self.emp_id, self.name, self.details)

employees = {}

for i in range(3):
    emp_id = input("ID: ")
    name = input("Name: ")
    dept = input("Department: ")
    salary = input("Salary: ")
    employees[emp_id] = Employee(emp_id, name, dept, salary)

for emp in employees.values():
    emp.show_details()
```

```python
import math
import random
from datetime import datetime

history = {}

while True:
    print("\n1. Basic Arithmetic")
    print("2. Scientific Calculation")
    print("3. Random Number")
    print("4. Store Result")
    print("5. View History")
    print("6. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        try:
            a = float(input("Enter first number: "))
            b = float(input("Enter second number: "))
            op = input("Enter (+,-,*,/): ")

            if op == "+":
                result = a + b
            elif op == "-":
                result = a - b
            elif op == "*":
                result = a * b
            elif op == "/":
                result = a / b
            else:
                print("Invalid Operator")
                continue

            print("Result:", result)

        except:
            print("Invalid Input")

    elif choice == "2":
        n = float(input("Enter number: "))
        result = math.sqrt(n)
        print("Square Root:", result)

    elif choice == "3":
        result = random.randint(1, 100)
        print("Random Number:", result)

    elif choice == "4":
        history[str(datetime.now())] = result
        print("Result Stored")

    elif choice == "5":
        print(history)

    elif choice == "6":
        print("Program End")
        break

    else:
        print("Invalid Choice")
```
